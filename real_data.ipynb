```python
import numpy as np
import pandas as pd
from fit_var import fit_var
from est_varma import estimate_varma
import matplotlib.pyplot as plt
import os
from est_model import estimate_model
```

```python
# Load the data
data = np.loadtxt("output.csv", delimiter=',', skiprows=0)  # skiprows=0 to not skip the header
print(data)
```

```python
moyenne = data.mean(axis=0)
sd = data.std(axis=0)
print(moyenne)
print(sd)
print(sd**2)
```

```python
# Center the data
def center_data(data):
    means = data.mean(axis=0)
    Y = data - means
    return Y
```

```python
centered_data = center_data(data)
print(centered_data)
```

```python
residuals, beta = fit_var(centered_data, 3)
print(beta)

# changement de forme pour adapter et comparer a l'output de la fonction de fit statsmodels
beta_reshaped = beta.T.reshape(2, 3, 2)
print(beta_reshaped)
```

```python
# Assuming `data` and `residuals` are already defined
p = 3
omega, Theta, Phi = estimate_varma(data, p, residuals)
print("Omega:\n", omega)
print("Theta:\n", Theta)
print("Phi:\n", Phi)
```

```python
A = -Theta
print(f"A = {A}")
B = Phi+Theta
print(f"B = {B}")
```

```python
#on cree un sous fichier pour y sauvegarder les replications
base_folder = "Real data"
os.makedirs(base_folder, exist_ok=True)
est = ([omega[0], A[0,0], A[0,1], B[0,0], B[0,1], omega[1], A[1,0], A[1,1], B[1,0], B[1,1]])

# Create DataFrame
estimstat = pd.DataFrame({
        'Parameter': [f'Parameter {i + 1}' for i in range(10)],
        'Mean': est,
        })

# Save DataFrame
estimstat_filename = f"estim_RD_MCO.csv"
estimstat_filepath = os.path.join(base_folder, estimstat_filename)
estimstat.to_csv(estimstat_filepath, index=False)
print(f"Saved statistics DataFrame to: {estimstat_filename}")
```
