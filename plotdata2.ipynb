```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Load your data
data = pd.read_csv('distsample.csv')
```

```python
data = pd.read_csv('databp.csv', header=None)
data.columns = ['Variable1', 'Variable2']

# Print the first few rows to verify
print(data.head())
```

```python
plt.figure(figsize=(16, 8))

# Plot Variable1 with a specific marker and transparency
plt.plot(data.index, data['Variable1'], label='Y1', marker='o', linestyle='-', color='b', alpha=0.7, markersize = 4)

# Plot Variable2 with a different marker and transparency
plt.plot(data.index, data['Variable2'], label='Y2', marker='s', linestyle='-', color='r', alpha=0.7, markersize = 4)

# Customize the plot
# plt.title('Graphe de la série temporelle bivariée')
# plt.xlabel('Points d\'Observation')
# plt.ylabel('Comptes')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
fig, axs = plt.subplots(2, 1, figsize=(12, 8), sharex=True)

# Plot Variable1
axs[0].plot(data.index, data['Variable1'], label='Y1', marker='o', linestyle='-', color='b', markersize= 4)
axs[0].set_title('Y1')
axs[0].set_ylabel('Counts')
axs[0].grid(True)

# Plot Variable2
axs[1].plot(data.index, data['Variable2'], label='Y2', marker='s', linestyle='-', color='r', markersize= 4)
axs[1].set_title('Y2')
axs[1].set_xlabel('Points d\'Observation')
axs[1].set_ylabel('Comptes')
axs[1].grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```
